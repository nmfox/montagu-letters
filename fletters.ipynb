```python
import nltk
import matplotlib
%matplotlib inline
```

```python
f = open('/Users/administrator/Desktop/DH_ENG/female.txt')
fletters = f.read()
```

```python
print(fletters[:30])
```

```python
fletters.count("woman")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stops = stopwords.words('english')
```

```python
fletters_tokens = word_tokenize(fletters)
```

```python
fletters_stops = []
for t in fletters_tokens:
    if t not in stops:
        fletters_stops.append(t)
```

```python
print(fletters_stops[:30])
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
fletters_clean = [wordnet_lemmatizer.lemmatize(t) for t in fletters_stops]
```

```python
print(len(fletters_clean))
print(len(set(fletters_clean)))
```
