```python
import nltk
import matplotlib
%matplotlib inline
```

```python
f = open('/Users/administrator/Desktop/DH_ENG/male.txt')
mletters = f.read()
```

```python
print(mletters[:30])
```

```python
mletters.count("pope")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stops = stopwords.words('english')
```

```python
mletters_tokens = word_tokenize(mletters)
```

```python
mletters_stops = []
for t in mletters_tokens:
    if t not in stops:
        mletters_stops.append(t)
```

```python
print(mletters_stops[:30])
```

```python
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()

mletters_clean = [wordnet_lemmatizer.lemmatize(t) for t in mletters_stops]

print(len(mletters_clean))
print(len(set(mletters_clean)))
```
